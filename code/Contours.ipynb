```python
%%bash
for year in {2012..2021}
do
    echo $year
    for contour in 1400 1500 1600 1700 1800 1900 2000
    do
        echo $contour
        gdal_contour -b 1 -amin from -amax to -fl -10000 $contour 10000 -p ../output/historic/$year.tif ../output/contours/$contour-$year.gpkg 
    done
done
```
